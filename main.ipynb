## __3.0 Modeling__ 

### __3.1 Data Preprocessing__

In our approach to processing and managing audio data for emotion recognition, we implemented a multi-step pipeline involving data loading, cleaning, preprocessing, feature extraction, and finally, data saving and splitting. Here’s a breakdown of the strategy and the reasoning behind each step:

#### 3.1.1 **Data Loading**

**Process**: We first load the audio data from a specified directory where the data is organized by emotion labels. For each emotion, we retrieve the audio files and store the audio data in an array.

**Reasoning**: 
- **Organization and Access**: We structured the data based on emotion categories to facilitate easy retrieval and management. This allows us to handle data more systematically and apply operations specific to each emotion.
- **Storage**: We use lists to store the raw audio data and corresponding labels, which makes it straightforward to process and access later.

#### 3.1.2 **Data Cleaning**

**Process**: After loading the audio files, we clean the data by applying noise reduction and trimming silence from the audio clips.

**Reasoning**:
- **Noise Reduction**: By reducing noise, we improve the quality of the audio, which can enhance the accuracy of subsequent feature extraction and emotion classification. This step helps in mitigating any distortions that might interfere with the analysis.
- **Silence Trimming**: Removing silence helps in focusing on the relevant parts of the audio, which can be crucial for training models that rely on the actual content of the audio rather than its length or non-informative parts.

#### 3.1.3 **Audio Preprocessing**

**Process**: We pad or truncate the audio data to a fixed length to ensure consistency across all samples.

**Reasoning**:
- **Consistency**: Uniform length for all audio samples is essential for training machine learning models, as it ensures that the input features are consistent. Padding short samples and truncating long ones standardizes the input size and simplifies further processing.

#### 3.1.4 **Feature Extraction**

**Process**: We extract Mel-Frequency Cepstral Coefficients (MFCCs) from the audio data. MFCCs are used to represent the audio signals in a way that captures the essential characteristics for emotion recognition.

**Reasoning**:
- **Feature Representation**: MFCCs are commonly used features in audio processing because they effectively capture the spectral properties of the audio signal, which are crucial for distinguishing between different emotions.
- **Dimensionality Reduction**: By averaging MFCCs across time, we reduce the dimensionality of the data while preserving the most relevant features, making it easier to train machine learning models.

#### 3.1.5 **Data Saving**

**Process**: We save the processed features and labels into CSV files. This step also involves splitting the data into training, validation, and test sets.

**Reasoning**:
- **Persistence**: Saving the processed data ensures that we can reuse it without having to repeat the preprocessing steps. This is time-efficient and allows us to maintain consistency across different runs.
- **Data Splitting**: Splitting the data into training, validation, and test sets is crucial for evaluating the performance of our models. It ensures that the models are trained on one subset, validated on another, and tested on a separate, unseen subset.

## **3.0 Modeling and Evaluation**

### **3.1 Data Preparation**

**Audio Preprocessing**

1. **Data Loading**: We began by loading audio files from a specified directory, where each folder corresponded to a different emotion. Audio files were loaded and stored in a list along with their associated labels. This step ensured that our dataset was properly organized and ready for further processing.

2. **Padding and Truncation**: We standardized the audio files to a target length by padding shorter files with zeros or truncating longer ones. This standardization ensured that all input samples had the same shape, which was necessary for the machine learning models we used.

**Feature Extraction**

1. **Extracting Acoustic Features**: After preprocessing, we extracted various features from the audio data, including:

   ##### 1. **MFCC_1 to MFCC_13 (Mel-Frequency Cepstral Coefficients)**
      - **MFCCs** are coefficients representing the short-term power spectrum of an audio signal, capturing its timbral texture.
      - **MFCC_1 to MFCC_13** represented the first 13 coefficients, commonly used for emotion recognition. 
      - **MFCC_1**: Captured the overall spectral slope.
      - **MFCC_2 to MFCC_13**: Captured more detailed characteristics of the spectral envelope.

   ##### 2. **Chroma_1 to Chroma_12**
      - **Chroma features** represented the 12 different pitch classes in Western music. They were useful for capturing harmonic content and identifying the musical key of a piece.
      - **Chroma_1 to Chroma_12** represented the energy distribution over these 12 pitch classes.

   ##### 3. **Spectral Contrast**
      - **Spectral contrast** referred to the difference in amplitude between peaks and valleys in the sound spectrum, aiding in distinguishing different timbres and textures.

   ##### 4. **Zero Crossing Rate**
      - The **Zero Crossing Rate (ZCR)** measured how frequently the signal changed sign from positive to negative or vice versa. High ZCR values indicated high-frequency content, often associated with noisy or percussive sounds.

   ##### 5. **RMS (Root Mean Square Energy)**
      - **RMS Energy** reflected the power or energy of the sound, capturing perceived loudness. Higher RMS values indicated louder audio, while lower values indicated softer sounds.

   ##### 6. **Spectral Centroid**
      - The **Spectral Centroid** indicated the center of mass of the spectrum and was associated with the perceived brightness of a sound. Higher values were typically associated with brighter sounds.

   ##### 7. **Spectral Bandwidth**
      - **Spectral Bandwidth** measured the width of the spectrum, indicating the range of frequencies present. A wider bandwidth corresponded to a more complex sound.

   ##### 8. **Spectral Roll-off**
      - The **Spectral Roll-off** point was the frequency below which a certain percentage of the total spectral energy lay, typically 85% or 95%. It helped to distinguish between voiced and unvoiced sounds or different types of noise.

### **3.2 Model Selection**

**Choosing Models**

1. **K-Nearest Neighbors (KNN)**: We used KNN, a simple algorithm that classifies data based on the majority class among its nearest neighbors. It was suitable for quick benchmarks and simple datasets.

2. **Random Forest**: We implemented Random Forest, an ensemble method that combined multiple decision trees to improve classification accuracy and robustness against overfitting.

3. **XGBoost**: We employed XGBoost, an advanced boosting algorithm that improved accuracy and performance through gradient boosting. It efficiently handled large datasets and was known for its high performance.

4. **CatBoost**: We included CatBoost, a gradient boosting method that efficiently handled categorical features and provided robust performance with less hyperparameter tuning.

5. **AdaBoost**: We incorporated AdaBoost, a boosting technique that improved the performance of weak learners by combining them into a strong classifier. It adjusted weights to focus on harder-to-classify examples.

6. **Stacking (Meta-model)**: We used stacking as a meta-modeling technique, combining multiple base models (like KNN, Random Forest, XGBoost, etc.) to enhance overall performance by leveraging their diverse strengths.

**Building the Model**

Each model was constructed based on its specific requirements. For instance, while KNN and Random Forest models were built and trained with standard hyperparameters, advanced models like XGBoost and CatBoost were tailored with specific configurations for optimal performance.

### **3.3 Training and Evaluation**

**Training Models**

1. **Model Training**: We fit each model to the training data, adjusting parameters to minimize prediction errors. 

**Parameter Tuning**

1. **Hyperparameter Optimization**:
   - **Objective**: Our goal was to find the optimal hyperparameters to enhance model performance and generalization.
   - **Methods**:
     - **Grid Search**: We systematically explored a specified set of hyperparameters using `GridSearchCV` to perform an exhaustive search.

**Preventing Overfitting and Underfitting**

1. **Overfitting Prevention**:
   - **Regularization**:

   - **Cross-Validation**:

     - **Stratified Cross-Validation**: We ensured balanced representation of each class in every fold for imbalanced datasets.
   - **Early Stopping**:
     - **Procedure**: We monitored validation loss during training and stopped training when performance ceased to improve to avoid overfitting.
   - **Ensemble Methods**:
     - **Bagging**: We used bagging techniques like Random Forest to reduce variance and overfitting.
     - **Boosting**: We focused on hard-to-classify examples with methods like AdaBoost and XGBoost to reduce bias and improve performance.

2. **Underfitting Prevention**:
   - **Model Complexity**:
     - **Increase Model Complexity**: We used more complex models or increased the number of features if the model was too simple to capture the underlying patterns.
   - **Feature Engineering**:
     - **Add Features**: We introduced additional relevant features or interactions between existing features to provide more information to the model.

   - **Training Duration**:
     - **Extend Training**: We increased the number of iterations to allow the model to learn more from the training data while monitoring for potential overfitting.
