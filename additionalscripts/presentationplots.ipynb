```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Function to add labels inside the pie charts
def add_labels(patches, texts, values):
    for i, (p, txt) in enumerate(zip(patches, texts)):
        percentage = values[i]
        txt.set_text(f'{percentage}%')
        p.set_edgecolor('white')
        x = p.get_center()[0]
        y = p.get_center()[1]
        p.set_text(f'{percentage}%')

# 1. No. of participants (242 total, 112 male, rest female)
plt.figure(figsize=(6, 6))
labels = ['Male', 'Female']
sizes = [112, 242-112]
colors = sns.color_palette("pastel")
patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, pctdistance=0.85)
plt.title('Participant Gender Distribution')
plt.axis('equal')
plt.show()

# 2. Number of recordings (1856 recorded, 1180 used, rest excluded)
plt.figure(figsize=(6, 6))
labels = ['Used', 'Processing/Excluded']
sizes = [1180, 1856-1180]
colors = sns.color_palette("pastel")
patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, pctdistance=0.85)
plt.title('Recording Status')
plt.axis('equal')
plt.show()

# 3. Percentage of recordings done online vs in person (88% online, rest in person)
plt.figure(figsize=(6, 6))
labels = ['Online', 'In Person']
sizes = [88, 100-88]
colors = sns.color_palette("pastel")
patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, pctdistance=0.85)
plt.title('Recording Method')
plt.axis('equal')
plt.show()

# 4. Distribution of emotions (anger, sad, surprised, calm, happy) equally distributed (bar chart)
plt.figure(figsize=(8, 6))
emotions = ['Anger', 'Sad', 'Surprised', 'Calm', 'Happy']
counts = [1180/5] * 5
sns.barplot(x=emotions, y=counts, palette="pastel")
plt.title('Emotion Distribution')
plt.xlabel('Emotions')
plt.ylabel('Number of Recordings')
plt.show()
```
